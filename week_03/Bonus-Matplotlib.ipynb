```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
```

## *First Experiment*

```python
x = np.linspace(0, 10, 50)
y0 = np.tan(x)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tan(-x)

plt.figure(figsize=(5, 10))
ax = plt.axes()
ax.set_facecolor("pink")
plt.plot(x, y0, color='cyan', linewidth=1.0)
plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--')

plt.figure(num=3, figsize=(10, 5))
ax = plt.axes()
ax.set_facecolor("cyan")
plt.plot(x, y2, color='pink', linewidth=5)
plt.plot(x, y1, color='violet', linewidth=1.0, linestyle='--')
plt.plot(x, -y1, color='violet', linewidth=1.0, linestyle='--')

plt.figure(figsize=(5, 10))
ax = plt.axes()
ax.set_facecolor("pink")
plt.plot(x, y3, color='cyan', linewidth=1.0)
plt.plot(x, -y2, color='red', linewidth=1.0, linestyle='--')


plt.show()
```

## *Second Test*

```python
n = 1024

X = np.random.normal(0, 1/2, n)
Y = np.random.normal(0, 1/2, n)
T = np.arctan2(Y, X)

ax = plt.axes()
ax.set_facecolor('black')
plt.scatter(X, Y, s=1, c=T, cmap='magma')

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.xticks(())
plt.yticks(()) # hide x and y ticks

plt.show()
```

```python
n = 2048

X = np.random.normal(0, 2, n)
Y = np.random.normal(0, 2, n)
T = np.arctan2(Y, X)

plt.scatter(X, Y, s=10000, c=T, alpha=0.1, edgecolor='white', cmap='inferno')
# colormaps:  'viridis', 'plasma', 'inferno', 'magma', 'cividis'

plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.xticks(())
plt.yticks(()) # hide x and y ticks

plt.show()
```
